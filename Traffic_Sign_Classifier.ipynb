```python
# Read the signnames.csv into a pandas dataframe
df = pd.read_csv('signnames.csv', index_col=False)
g_SIGN_NAMES = df.SignName.values

# Utility function to return the counts of each label in the data
def get_counts_of_sign_name(y, dataFrame=df):
    """
    param y: List [1, 0, 1, 1, 1, 0, 0, 1]
    param dataFrame: pandas dataframe ["ClassId": {0, 1}, "SignName": {"Class0", "Class1"}]
    returns counts_of_sign_name :dict {"Class0": 3, ""Class1": 5} 
    """
    counts_of_sign_name = {}
    for label in y:
        sign_name = dataFrame.loc[label, 'SignName']
        if sign_name not in counts_of_sign_name:
            counts_of_sign_name[sign_name] = 0
        counts_of_sign_name[sign_name] += 1
    return counts_of_sign_name
```

We will now explore whether the frequencies of the labels in the validation data matches the frequencies of labels in training data. For this, we compute perform a chi-squared test.

## Augmenting training data

```python
def random_brightness(image, ratio):
    """
    Randomly adjust brightness of the image.
    """
    # HSV (Hue, Saturation, Value) is also called HSB ('B' for Brightness).
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    brightness = np.float64(hsv[:, :, 2])
    brightness = brightness * (1.0 + np.random.uniform(-ratio, ratio))
    brightness[brightness>255] = 255
    brightness[brightness<0] = 0
    hsv[:, :, 2] = brightness
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)


def random_rotation(image, angle):
    """
    Randomly rotate the image
    """
    if angle == 0:
        return image
    angle = np.random.uniform(-angle, angle)
    rows, cols = image.shape[:2]
    size = cols, rows
    center = cols/2, rows/2
    scale = 1.0
    rotation = cv2.getRotationMatrix2D(center, angle, scale)
    return cv2.warpAffine(image, rotation, size)


def random_translation(image, translation):
    """
    Randomly move the image
    """
    if translation == 0:
        return 0
    rows, cols = image.shape[:2]
    size = cols, rows
    x = np.random.uniform(-translation, translation)
    y = np.random.uniform(-translation, translation)
    trans = np.float32([[1,0,x],[0,1,y]])
    return cv2.warpAffine(image, trans, size)


def random_shear(image, shear):
    """
    Randomly distort the image
    """
    if shear == 0:
        return image
    rows, cols = image.shape[:2]
    size = cols, rows
    left, right, top, bottom = shear, cols - shear, shear, rows - shear
    dx = np.random.uniform(-shear, shear)
    dy = np.random.uniform(-shear, shear)
    p1 = np.float32([[left   , top],[right   , top   ],[left, bottom]])
    p2 = np.float32([[left+dx, top],[right+dx, top+dy],[left, bottom+dy]])
    move = cv2.getAffineTransform(p1,p2)
    return cv2.warpAffine(image, move, size)
    
    
def augment_image(image, brightness, angle, translation, shear):
    image = random_brightness(image, brightness)
    image = random_rotation(image, angle)
    image = random_translation(image, translation)
    image = random_shear(image, shear)
    return image
```

### Model Architecture for LeNet

Layer # | Layer Type | Output Shape
:---:| :--- | :---
**Input**||32x32x3 (3 because there are three color channels)
**1**| Convolutional | 28x28x24
**1** | Activation (Relu) | 28x28x24
**1**| Pooling (Max) | 14x14x24
**2**| Convolutional | 10x10x64
**2** | Activation (Relu) | 10x10x64
**2**| Pooling (Max) | 5x5x64
| Flatten | 1600
**3**| Fully Connected | 480
**3**| Activation (Relu) | 480
**4**| Fully Connected | 43 (The number of unique labes in the training data)
**Output**| Logits | 43

```python
import tensorflow as tf
from tensorflow.contrib.layers import flatten
EPOCHS = 50
BATCH_SIZE = 50
g_number_of_classes = 43

tf.reset_default_graph()
def LeNet(x, dropout):
    
    # Statistics of normal distribution to initialize weights of the convolutional neural nets. 
    mu = 0
    sigma = 0.1
    
    # Layer 1
    conv1_W = tf.Variable(tf.truncated_normal([5, 5, 3, 24], mean=mu, stddev=sigma))
    conv1_b = tf.Variable(tf.zeros(24))
    l1 = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID')
    l1 = tf.nn.bias_add(l1, conv1_b)
    l1 = tf.nn.relu(l1)
    l1 = tf.nn.max_pool(l1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    # Layer 2
    conv2_W = tf.Variable(tf.truncated_normal([5, 5, 24, 64], mean=mu, stddev=sigma))
    conv2_b = tf.Variable(tf.zeros(64))
    l2 = tf.nn.conv2d(l1, conv2_W, strides=[1, 1, 1, 1], padding='VALID')
    l2 = tf.nn.bias_add(l2, conv2_b)
    l2 = tf.nn.relu(l2)
    l2 = tf.nn.max_pool(l2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    flat_l2 = flatten(l2)
    
    # Layer 3
    in_neurons = 5*5*64
    out_neurons = 480
    fc3_W = tf.Variable(tf.random_normal([in_neurons, out_neurons], mean=mu, stddev=np.sqrt(2.0/in_neurons)))
    fc3_b = tf.Variable(tf.zeros(out_neurons))
    l3 = tf.add(tf.matmul(flat_l2, fc3_W), fc3_b)
    l3 = tf.nn.relu(l3)
    l3 = tf.nn.dropout(l3, dropout)
    
    """
    # Layer 4
    in_neurons = out_neurons
    out_neurons = 240
    fc4_W = tf.Variable(tf.random_normal([in_neurons, out_neurons], mean=mu, stddev=np.sqrt(2.0/in_neurons)))
    fc4_b = tf.Variable(tf.zeros(240))
    l4 = tf.add(tf.matmul(l3, fc4_W), fc4_b)
    l4 = tf.nn.relu(l4)
    l4 = tf.nn.dropout(l4, dropout)
    """
    l4 = l3 # Remove this if you want l4 to be different!
    
    # Layer 5
    in_neurons = out_neurons
    out_neurons = g_number_of_classes
    fc5_W = tf.Variable(tf.random_normal([in_neurons, out_neurons], 
                                         mean=mu, 
                                         stddev=np.sqrt(2.0/in_neurons)))
    fc5_b = tf.Variable(tf.zeros(g_number_of_classes))
    logits = tf.add(tf.matmul(l4, fc5_W), fc5_b)
    
    return logits
```

```python
keep_prob = tf.placeholder(tf.float32)
x = tf.placeholder(tf.float32, (None, 32, 32, 3))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, g_number_of_classes)
```

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation

```python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.
```

```python
# Creating the training operation

learning_rate = 0.0005

logits = LeNet(x, keep_prob)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
training_operation = optimizer.minimize(loss)
```

```python
# Evaluating the accuracy

correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
```

```python
# Plotting training and validation accuracy
def show_learning_curve(training_accuracy, validation_accuracy):
    plt.plot(range(len(training_accuracy)), training_accuracy)
    plt.plot(range(len(validation_accuracy)), validation_accuracy)
    plt.axis((0, max(len(training_accuracy), len(validation_accuracy)) - 1, 0.6, 1.0))
    plt.show()
```

```python
# Training

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    training_accuracy = [0]
    validation_accuracy = [0]
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 0.70})
        
        training_accuracy.append(evaluate(X_train, y_train))
        validation_accuracy.append(evaluate(X_valid, y_valid))
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy[-1]))
        print("Training Accuracy = {:.3f}".format(training_accuracy[-1]))
        print()
        
    show_learning_curve(training_accuracy, validation_accuracy)
        
    saver.save(sess, './lenet')
    print("Model saved")
```
